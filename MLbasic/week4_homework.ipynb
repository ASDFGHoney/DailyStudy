```python
import torch
from numpy import dot 
from numpy.linalg import norm 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visualize(distance, name):
    # Sort the distances
    idx = np.argsort(distance) # -> 기준점과 거리가 가까운 순으로 정렬
    # return the indices of K nearest neighbor
    neighborIndex = idx[:k] # 기준점과 가까운 k개의 데이터 추출

    plt.plot(rand_data[0, :], rand_data[1, :], "ok", label="Data")
    plt.plot(x[0, 0], x[1,0], 'or', label='Query')
    plt.plot(rand_data[0, neighborIndex], rand_data[1,neighborIndex], 'o'
            , markerfacecolor='None', markersize =12, markeredgewidth=0.5, label='KNN')
    plt.legend(loc='upper right')
    plt.title('KNN search ('+name+')')
    plt.show()
```

```python
k=10
rand_data = np.random.rand(2, 100)

dataLen = rand_data.shape[1]

if k < dataLen:
    k=k
else:
    k = dataLen
```

```python
# 이 셀을 실행 할 때 마다 같은 데이터 범위에서 기준점 변경
x = np.random.rand(2,1)

#기준점과 데이터간 거리(distance) 계산
euclidean_distance = np.sqrt(((rand_data-x)**2).sum(axis=0))
cosine_distance = 1 - (dot(np.transpose(x), rand_data)/(norm(rand_data)*norm(x)))[0]
manhattan_distance = abs(rand_data - x).sum(axis=0)

# 시각화
visualize(euclidean_distance, 'euclidean_distance')
visualize(cosine_distance, 'cosine_distance')
visualize(manhattan_distance, 'manhattan_distance')
```
