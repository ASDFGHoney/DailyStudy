# KNN 알고리즘 구현

```python
k=10
rand_data = np.random.rand(2, 100)
x = np.random.rand(2,1)
```

```python
A = np.random.rand(2,2)
print("행렬 A : \n", A)
B = np.random.rand()
print("행렬 B : \n", B)
```

```python
# 데이터 기준점 생성
dataLen = rand_data.shape[1]

if k < dataLen:
    k=k
else:
    k = dataLen
```

```python
#기준점과 데이터간 거리(distance) 계산
distance = np.sqrt(((rand_data-x)**2).sum(axis=0))
# Sort the distances
idx = np.argsort(distance) # -> 기준점과 거리가 가까운 순으로 정렬
# return the indices of K nearest neighbor
neighborIndex = idx[:k] # 기준점과 가까운 k개의 데이터 추출
```

```python
plt.plot(rand_data[0, :], rand_data[1, :], "ok", label="Data")
plt.plot(x[0, 0], x[1,0], 'or', label='Query')
plt.plot(rand_data[0, neighborIndex], rand_data[1,neighborIndex], 'o'
         , markerfacecolor='None', markersize =12, markeredgewidth=0.5, label='KNN')
plt.legend(loc='upper right')
plt.title('KNN search')
plt.show()
```
